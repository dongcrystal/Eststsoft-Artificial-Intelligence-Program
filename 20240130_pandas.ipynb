### 1.1.데이터프레임 이해

#### 1.1.1.데이터프레임이란


- Pandas 사용 목적이 '**데이터프레임**'을 사용하기 위한 목적으로 봐도 됩니다.
- 데이터를 처리, 조회, 분석하는 가장 효율적인 방법이 데이터프레임을 사용하는 것입니다.
- 일반적으로 접하게 되는 **테이블** 형태, **엑셀** 형태로 생각하시면 됩니다.
- 직접 만들 수 있으나 보통은 **csv 파일**, **엑셀 파일** 또는 DB에서 읽어옵니다.

### 1.2.데이터프레임 직접 만들기

- **pd.DataFrame()** 함수를 사용해 데이터프레임을 직접 만들 수 있습니다.
- 대부분 리스트, 딕셔너리, Numpy 배열로부터 데이터프레임을 만듭니다.
- 데이터프레임을 만들기 위해서는 다음 세 가지를 위한 데이터가 필요합니다.
    - 데이터
    - 열 이름
    - 인덱스 이름
- 열 이름을 지정하지 않으면 **열 번호**에 기반한 정수(0, 1, 2...)가 열 이름이 됩니다.
- 인덱스 이름을 지정하지 않으면 **행 번호**에 기반한 정수(0, 1, 2...)가 인덱스 이름이 됩니다.

#### 1.2.1.라이브러리 불러오기

- 데이터프레임을 사용하기 위해서 pandas 라이브러리를 pd 별칭을 주어 불러옵니다.

```python
# 라이브러리 불러오기
import pandas as pd
```

#### 1.2.2.딕셔너리로 만들기

- 딕셔너리로 데이터프레임을 만들면 딕셔너리의 **키**가 **열 이름**이 됩니다.
- 인덱스를 지정하지 않으면 행 번호가 인덱스가 됩니다.

```python
# 딕셔너리 만들기
d = {'Name': ['Gildong', 'Sarang', 'Jiemae', 'Yeoin'],
        'Level': ['Gold', 'Bronze', 'Silver', 'Gold'],
        'Score': [56000, 23000, 44000, 52000]}

# 확인
d
```

```python
# 데이터프레임 만들기
df = pd.DataFrame(d)

# 확인
df
```

- **read_csv()** 함수를 사용해서 CSV 파일에서 데이터를 읽어옵니다.

**[주요옵션]**
- **usecols** : 불러올 칼럼 지정(칼럼이름을 알고 있어야 함)
- sep: 구분자 지정(기본값 = 콤마)
- header: 헤더가 될 행 번호 지정(기본값 = 0)
- index_col: 인덱스 열 지정(기본값 = False)
- names: 열 이름으로 사용할 문자열 리스트
- skiprows: 처음 몇 줄을 무시할 것인지 지정, 리스트로 지정 가능

### 2.4.기본 집계

- 데이터를 좀더 이해하기 위해 고유값, 합, 평균, 최댓값, 최솟값 등을 확인합니다.

#### 2.4.1.고유값 확인

- 범주형 열(열이 가진 값이 일정한 값인 경우, 성별, 등급 등)인지 확인할 때 사용합니다.

**1) 고유값 확인**

- unique() 메소드로 고유값을 확인하며, 결과값은 배열 형태가 됩니다.

```python
# ShelveLoc 열 고유값 확인
print(temp['ShelveLoc'].unique())
```

**2) 고유값과 개수 확인**

- value_counts() 메소드로 고유값과 그 개수를 확인하며, 결괏값은 시리즈 형태가 됩니다.

```python
# ShelveLoc 열 고유값 개수 확인
print(temp['ShelveLoc'].value_counts())
```

```python
# Education 열 고유값 개수 확인
print(temp['Education'].value_counts())
```

#### 2.4.2.기본 통계 메소드 사용

- 데이터를 1차 집계 한 후 분석을 진행하는 경우가 많으므로 필히 알아두어야 할 내용입니다.

```python
# 'Age', 'Income' 열 평균값 확인
print(temp[['Age', 'Income']].mean())
```

```python
# ''Age', 'Income' 열 중앙값 확인
print(temp[['Age', 'Income']].median())
```

### 3.3.조건으로 조회

- **df.loc[조건]** 형태로 조건을 지정해 조건에 만족하는 데이터만 조회할 수 있습니다.
- 우선 조건이 제대로 판단이 되는지 확인한 후 그 **조건을 대 괄호 안에** 넣으면 됩니다.

**1) 단일 조건 조회**

```python
# 조건절(조건문)의 결과는 True, False
data['DistanceFromHome'] > 10
```

```python
# DistanceFromHome이 퇴사에 영향을 미칠까
# 열 값이 10 보다 큰 행 조회. 열값을 바꿔가며 이직여부 확인
data.loc[data['DistanceFromHome'] > 10]
```

```python
# 이직한 사람을 조건 조회하여 이직에 영향을 미칠 변수를 생각해 봅시다
data.loc[data['Attrition'] == 1 ]
```

**2) 여러 조건 조회**

- [ ]안에 조건을 여러개 연결할 때 **and와 or 대신에 &와 |**를 사용해야 합니다.
- 그리고 각 조건들은 **(조건1) & (조건2)** 형태로 **괄호**로 묶어야 합니다.

```python
# 이직에 영향을 미칠것으로 예상되는 변수를
# and로 여러 조건 연결하여 찾아봅니다
data.loc[(data['PercentSalaryHike'] < 20) & (data['TotalWorkingYears'] < 10)]
```

```python
# or 조건 : |
data.loc[(data['PercentSalaryHike'] < 15) | (data['TotalWorkingYears'] < 3)]
```

**3) 편리한 isin(), between() 메소드**

- **isin([값1, 값2,..., 값n])**: 값1 또는 값2 또는...값n인 데이터만 조회합니다.
- 주의 isin(**리스트**) 값들을 리스트 형태로 입력해야 합니다.

```python
# 값 나열
data.loc[data['JobSatisfaction'].isin([1, 4])]
```

- 위 구문은 다음과 같은 의미를 갖습니다.

```python
# or 조건
data.loc[(data['JobSatisfaction'] == 1) | (data['JobSatisfaction'] == 4)]
```

- **between(값1, 값2)**: 값1 ~ 값2까지 범위안의 데이터만 조회합니다.

```python
# 범위 지정
data.loc[data['Age'].between(18, 25)]
```

- 위 구문은 다음과 같은 의미를 갖습니다.

```python
# and 조건
data.loc[(data['Age'] >= 18) & (data['Age'] <= 25)]
```

**4) 조건을 만족하는 행의 일부 열 조회**

- **df.loc[조건, ['열 이름1', '열 이름2',...]]** 형태로 조회할 열을 리스트로 지정합니다. ==> 2차원, 데이터프레임형태로 조회

```python
# 조건에 맞는 하나의 열 조회
data.loc[data['MonthlyIncome'] > 7000, ['Age']]
```

```python
# 조건에 맞는 여러 열 조회
data.loc[data['MonthlyIncome'] > 7000, ['Age', 'MaritalStatus', 'TotalWorkingYears']]
```
